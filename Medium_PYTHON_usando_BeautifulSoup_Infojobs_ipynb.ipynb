```
from bs4 import BeautifulSoup
import re
import pandas as pd
from urllib.request import urlopen
```

```
empregos=[]

for n in range(1000): #Informar a qtd de páginas qe deseja pesquisar
  url = "https://www.infojobs.com.br/empregos.aspx?Page="+str(n)
  pag_empregos = urlopen(url)
  bs = BeautifulSoup(pag_empregos, 'lxml')
  for emprego in bs.find_all("div", attrs={"class": "vaga"}):
    dados_emprego = []
    dados_emprego.append(url)
    dados_emprego.append(emprego.a['title'])
    dados_emprego.append(emprego.a['href'])
    empregos.append(dados_emprego)
```

```
empregos_df = pd.DataFrame(empregos, columns=['pagina','vaga','url'])
```

```
bd_vagas=[]
for vaga in empregos_df['url']:
  url_vaga = vaga
  pag_vagas = urlopen(url_vaga)
  #print(pag_vagas)
  bs_vaga = BeautifulSoup(pag_vagas, 'html.parser')
  dados_vaga= []
  try:
    dados_vaga.append(bs_vaga.find(id='ctl00_phMasterPage_cVacancySummary_litVacancyTitle').text) #Titulo
    dados_vaga.append(bs_vaga.find(id='ctl00_phMasterPage_cVacancySummary_aCompany').text)        #Empresa
    dados_vaga.append(bs_vaga.find(id='ctl00_phMasterPage_cVacancySummary_litWorkingHours').text) #Carga horária
    dados_vaga.append(bs_vaga.find(id='ctl00_phMasterPage_cVacancySummary_litContractType').text) #Tipo de contrato
    dados_vaga.append(bs_vaga.find(id='ctl00_phMasterPage_cVacancySummary_litSalary').text)       #Salário
    dados_vaga.append(bs_vaga.find(id='ctl00_phMasterPage_cVacancySummary_litLocation').text)     #Localização
    dados_vaga.append(bs_vaga.find('ol', attrs={"class":'descriptionItems'}).text)                #Descricao
    bd_vagas.append(dados_vaga)
  except:
    print('erro')
```

```
bd_vagas_df = pd.DataFrame(bd_vagas, columns=['titulo','empresa','cargahoraria','tipo_contrato','salario', 'localizacao','descricao'])
```
